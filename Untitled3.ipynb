```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/Netflix Userbase.csv")
data.head()
data.describe()
```

```python
sns.countplot(x='Gender', data=data)
plt.title('Gender distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='Gender', data=data,hue='Gender',color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
subscription_counts = data['Subscription Type'].value_counts()
subscription_counts
```

```python
subscription_counts = data['Subscription Type'].value_counts()
plt.pie(subscription_counts, labels=subscription_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Subscription_types')
plt.show()
```

```python
sns.boxplot(x='Subscription Type', y='Monthly Revenue', data=data)
plt.title('Monthly Revenue by Subscription Type')
plt.xlabel('Subscription Type')
plt.ylabel('Monthly Revenue')
plt.show()
```

```python
corr_matrix = data[['Monthly Revenue', 'Age']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='crest')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.countplot(y='Country', data=data)
plt.title('Country Disribution')
plt.xlabel('Country')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Gender', y='Monthly Revenue', data=data)
plt.title('Monthly Revenue by Gender')
plt.xlabel('Gender')
plt.ylabel('Monthly Revenue')
plt.show()
```
